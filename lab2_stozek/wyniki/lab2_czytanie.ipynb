```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
import os
import plotly.graph_objects as go
```

```python
g = 9.81
alpha = 0.5
m0 = 1.
```

```python
def czytaj(file_path):
    with open(file_path, 'r') as file:
        data = []
        for line in file:
            values = list(map(float, line.split()))
            data.append(values)
        data = np.array(data)
        data = data.T
        return data
```

```python
data = []
folder = ''
for i in range(5):
    plik = f'lab2_stozek_{i}.txt'
    sciezka = os.path.join(folder, plik)
    if os.path.exists(sciezka):
        print(f"Otwieranie {plik}")
        data.append(czytaj(sciezka))
    else:
        print(f"nie istnieje.")
```

```python
t = []
phi = []
z = []
omega = []
vz = []
E = []
for i in range(2):
    t.append(data[i][0])
    phi.append(data[i][1])
    z.append(data[i][2])
    omega.append(data[i][3])
    vz.append(data[i][4])
    E.append(data[i][5])
# E2 = 1/2*((m.tan(alpha)*z[0]*omega[0])**2 + (vz[0]/m.cos(alpha))**2) + g*z[0]*m.sin(alpha)*(1-np.cos(phi[0]))
```

```python
fs=15

phi_value = [4,45,90,135,175]
for i in range(2):
    label = fr'$\phi = {phi_value[i]}^\circ$'
    plt.plot(t[i],phi[i], label=label)
plt.xlabel('t', fontsize=fs)
plt.ylabel(r'$\phi$', fontsize=fs)
plt.legend(loc='upper right')
plt.show()

phi_value = [4,45,90,135,175]
for i in range(2):
    label = fr'$\phi = {phi_value[i]}^\circ$'
    plt.plot(t[i],z[i], label=label)
plt.xlabel('t', fontsize=fs)
plt.ylabel(r'$z$', fontsize=fs)
plt.legend(loc='upper right')
plt.show()

phi_value = [4,45,90,135,175]
for i in range(2):
    label = fr'$\phi = {phi_value[i]}^\circ$'
    plt.plot(t[i],omega[i], label=label)
plt.xlabel('t', fontsize=fs)
plt.ylabel(r'$omega$', fontsize=fs)
plt.legend(loc='upper right')
plt.show()

phi_value = [4,45,90,135,175]
for i in range(2):
    label = fr'$\phi = {phi_value[i]}^\circ$'
    plt.plot(t[i],vz[i], label=label)
plt.xlabel('t', fontsize=fs)
plt.ylabel(r'$v_z$', fontsize=fs)
plt.legend(loc='upper right')
plt.show()

phi_value = [4,45,90,135,175]
for i in range(2):
    label = fr'$\phi = {phi_value[i]}^\circ$'
    plt.plot(t[i],E[i], label=label)
plt.xlabel(r'$\phi$', fontsize=fs)
plt.ylabel(r'$\frac{d\phi}{dt}$', fontsize=fs)
plt.legend(loc='upper right')
plt.show()
```

```python
ro = []
x = []
y = []

for i in range(2):
    ro.append(z[i]*m.tan(alpha))
    x.append(ro[i]*np.cos(phi[i]))
    y.append(ro[i]*np.sin(phi[i]))
```

```python
x2 = []
y2 = []
z2 = []

theta = m.pi/2 -alpha

for i in range(2):
    x2.append(x[i]*m.cos(theta) + z[i]*m.sin(theta))
    y2.append(y[i])
    z2.append(-x[i]*m.sin(theta) + z[i]*m.cos(theta))
```

```python
plt.plot(x2[0], y2[0])
plt.show()

plt.plot(x2[0], z2[0])
plt.show()

plt.plot(y2[0], z2[0])
plt.show()
```
