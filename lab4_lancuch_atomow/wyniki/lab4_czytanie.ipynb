```python
import numpy as np
import matplotlib.pyplot as plt
import math as m
import os
import plotly.graph_objects as go
```

```python
alpha = 1.0
N = 50
m = 1.0
```

```python
def czytaj(file_path):
    with open(file_path, 'r') as file:
        data = []
        for line in file:
            values = list(map(float, line.split()))
            data.append(values)
        data = np.array(data)
        data = data.T
        return data
```

```python
data = []
folder = ''
for i in range(7):
    plik = f'lab4_atomy_{i}.txt'
    sciezka = os.path.join(folder, plik)
    if os.path.exists(sciezka):
        print(f"Otwieranie {plik}")
        data.append(czytaj(sciezka))
    else:
        print(f"nie istnieje.")
```

```python
t = []
T = []
U = []
x = []
v = []

for i in range(7):
    t.append(data[i][0])
    T.append(data[i][-2])
    U.append(data[i][-1])
    x.append(data[i][1:N+2])
    v.append(data[i][N+2:-2])
# E2 = 1/2*((m.tan(alpha)*z[0]*omega[0])**2 + (vz[0]/m.cos(alpha))**2) + g*z[0]*m.sin(alpha)*(1-np.cos(phi[0]))
```

```python
E = []
for i in range(7) :
    E.append(U[i]+T[i])
```
